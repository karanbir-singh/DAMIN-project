```python
from langdetect import detect
import pandas as pd
```

```python
df_spotify = pd.read_csv("../data/spotify.csv")
```

```python
languages = set()
for i, track_name in enumerate(df_spotify['track_name']):
    try:   
        languages.add(detect(track_name))
    except:
        with open("../data/strange_track_names", "a", encoding="utf-8") as f:
            f.write(f"{track_name}\n")
languages
```

```python
languages = set()
for album_name in df_spotify['album_name']:
    try:   
        languages.add(detect(album_name))
    except:
        with open("../data/strange_album_names", "a", encoding="utf-8") as f:
            f.write(f"{album_name}\n")
```
